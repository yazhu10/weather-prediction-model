```python
numeric_col = df.columns[df.dtypes!='object']
catgoric_col = df.columns[df.dtypes =='object']
```

```python
print(f'num_col:{numeric_col}')
print(f'catgoric_col:{catgoric_col}')
```

```python
#Fill missing values for num columns with the mean
df[numeric_col] = df[numeric_col].apply(lambda col : col.fillna(col.median()))

#Fill missing values for cat columns with the mode
df[catgoric_col] = df[catgoric_col].apply(lambda col: col.fillna(col.mode))
```

```python
#remove outlayer using  IQR method
q1 = df[numeric_col].quantile(0.10)
q2 = df[numeric_col].quantile(0.80)
q3 = q2 - q1

data = df [~((df[numeric_col] < (q1 -1.5 * q3)) | (df[numeric_col] > (q2 +1.5 * q3))).any(axis=1)]
q3
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.6, random_state= 40)
```

```python
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
#MODEL TRAINING
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def evaluate_model(true,pred):
        f1_score = f1_score(true,pred)
        precision_score = precision_score(true,pred)
        recall_score = recall_score(true,pred)

        return f1_score, precision_score,recall_score

```

```python
models = {'RandomForestRegressor':RandomForestRegressor(),
          'LogisticRegression':LogisticRegression(),
          'DecisionTreeClassifier':DecisionTreeClassifier(),
          'DecisionTreeRegressor':DecisionTreeRegressor()}
          
```

```python
list(models)
```

```python
#for i in range (len(list(models))):
#model = list(models.values())[i]
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    
    
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f'accuracy: {accuracy}')
print(f'precision:{accuracy}')
print(f'recall:{recall}')
print(f'f1:{f1}')
```
